## Overview

Every variable or other object in python has a type. This is the kind of
information which is stored insice the variable or object Here is an
overview over the most important simple data types.

| Type       | Example   | Description           |
|------------|-----------|-----------------------|
| `int`      | `5`       | Integer               |
| `float`    | `3.14`    | Floating-point number |
| `complex`  | `2+3j`    | Complex number        |
| `bool`     | `True`    | Boolean value         |
| `str`      | `"hello"` | Text string           |
| `NoneType` | `None`    | No value / null value |

You can find about the type of any expression `x` in Python using
`type(x)`.

## Integers

Integers (int) are numbers …,-2,-1,0,1,2,… In most programming
languages, like `C++`, there is a minimal and a maximal possible value
for an int. Python, however, does not have this limitation. Only the
amount of available memory restricts the numbers which can be stored in
an int. With int, we can compute as usual.

```python
x = 7
y = 2
print("x + y = ", x + y)
print("x * y =  ", x * y)
# x to the poewr of y
print("x ** y = ", x ** y)
# x divided by y (which gives a float)
print("x / y = ", x / y)
# integer-valued division of x by y 
print("x // y = ", x // y)
# x modulo y
print("x % y = ", x % y)
```

## Floats

Computing with float and int gives an int.

Unlike ints, floats do have a maximal and minimal element. The reason is
that every float is stored in 64 bits, one for the sign, 11 for the
exponent and 52 for the mantissa. As a result, floats range from about
\$ ~10^{-324}\$ to $10^{308}$.

```python
y = 2
x + y
x * y
x ** y
x % y
0.1 + 0.2
```

Special floats are `inf`, `-inf` and `nan` (not a number). Let us see
them in action:

```python
(10. ** 308) * 10
(10 ** 308) * 10
(10 ** (-323)) / 10
(10. ** 308) * 10 - (10. ** 308) * 10
```

You can transform ints to floats by using

```python
float(0)
int(0.0)
int(0.5)
float("inf")
float("nan")
```
