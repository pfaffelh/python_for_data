# Python for Data Analysis

These are the materials which are used for a course on *Python for Data
Analysis* at the University of Freiburg in summer 2026. It does not
assume any prior knowledge of python (although this might help). The
goal of the course is some literacy in python, including basic knowledge

regression.

## Learning goals

- Install `git` and clone the course repository from `github`
- Install an Integrated Development Environment (IDE), if needed
- Learn about text and cells in a jupyter notebook

## Your python environment

We will assume that python is installed on your system. These course
notes are written with python 3.13, but other versions might work as
well. In addition to plan `python`, we will need some libraries. Which
ones we need (including their versions) is written in the file
`requirements.txt` in the main folder of these notes. We recomment that
you do the following:

1.  Create a virtual environment using `python -m venv venv` or
    `python3 -m venv venv`.
2.  Activate this environment by `source venv/bin/activate`,
3.  Install the libraries using `pip3 install -r requirements.txt` (or
    `pip install -r requirements.txt`).

General note: We avoid repeating the distinction between the commands
`python` and `python3`, and will simply write `python3` in the sequel.

## Starting a Jupyter notebook

There are basically three ways `python` commands can be run:

1.  In interactive mode on the command line by typing `python3` in your
    shell,
2.  In `python`-scripts, usually ending with `myScript.py`. You can run
    these using `python3 myScript.py`.
3.  Jupyter notebooks have the opportunity to mix usual text with
    `python` commands. We will use these in this course.

For Jupyter notebooks, there are basically two ways to run them:

1.  Start the Jupyter server using `jupyter lab` (hopefully, it is
    already installed), which opens your browser where you can edit and
    save what you have done;
2.  Use `vscode` as you IDE and work from there; see below.

## vscode

When programming, we need to type code. This is possible in a code
editor, and best done in a more sophisticated environment such as
[`vscode`](https://code.visualstudio.com/). It is best to acticvate your
vertual environment using `source venv/bin/activate` before starting
vscode. In order to start vscode, go to the main folder of the course
notes, and type `code .` vscode can be used for all three ways to write
Python commands:

1.  Use `Terminal -> New Terminal` in order to open a terminal, type
    `python3` and you are in interactive mode,
2.  Write a Python script and run it using the opoened terminal,
3.  Click on some Jupyther notebook, and vscode starts the Jupyter
    server for you.
