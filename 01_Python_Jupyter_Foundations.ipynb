## Functions and docstrings

```python
def factorial(n): 
    """Compute n! for a nonnegative integer n."""    
    result = 1    
    for k in range(2, n + 1):        
        result *= k    
        return result

[factorial(n) for n in range(8)]
```

```python
def arithmetic_mean(values):    
    return sum(values) / len(values)
arithmetic_mean([1, 4, 9, 16])
```

```python
def is_prime(n):    
    if n < 2:        
        return False    
    for k in range(2, int(n**0.5) + 1):
        if n % k == 0:
            return False
    return True

[p for p in range(2, 50) if is_prime(p)]
```

## Mini numerical experiment

```python
# Approximate pi via Monte Carlo simulation
import numpy as np
rng = np.random.default_rng(42)
n = 10_000
points = rng.uniform(-1, 1, size=(n, 2))
inside = np.sum(points[:, 0]**2 + points[:, 1]**2 <= 1)
pi_estimate = 4 * inside / n
pi_estimate
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 6))
sample = points[:800]
mask = sample[:, 0]**2 + sample[:, 1]**2 <= 1
plt.scatter(sample[mask, 0], sample[mask, 1], s=10, label="inside")
plt.scatter(sample[~mask, 0], sample[~mask, 1], s=10, label="outside")
plt.gca().set_aspect("equal")
plt.legend()
plt.title("Monte Carlo approximation of π")
plt.show()
```
